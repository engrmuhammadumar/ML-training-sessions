```python
import math

x = [i * 0.1 for i in range(0, 63)]      # 0.0, 0.1, 0.2, ... up to ~6.2
y_sin = [math.sin(value) for value in x]
y_cos = [math.cos(value) for value in x]

plt.figure(figsize=(8, 4))
plt.plot(x, y_sin, label="sin(x)", color="crimson", linewidth=2)
plt.plot(x, y_cos, label="cos(x)", color="navy", linewidth=2, linestyle="--")
plt.title("Sine and Cosine Waves")
plt.xlabel("x")
plt.ylabel("value")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Step 1: store the data in a dictionary
report = {
    "Math": 88,
    "Science": 95,
    "English": 72,
    "History": 80,
    "Computer": 99
}

# Step 2: a function that turns a number into a letter grade
def letter_grade(mark):
    if mark >= 90:
        return "A"
    elif mark >= 80:
        return "B"
    elif mark >= 70:
        return "C"
    else:
        return "F"

# Step 3: print a neat report using a loop
print("=" * 32)
print("        REPORT CARD")
print("=" * 32)
for subject, mark in report.items():
    print(f"{subject:<12} {mark:>3}   Grade: {letter_grade(mark)}")
print("=" * 32)

# Step 4: show some statistics
marks = list(report.values())
average = sum(marks) / len(marks)
print(f"Average mark : {average:.1f}")
print(f"Best subject : {max(report, key=report.get)}")
print("=" * 32)
```

```python
subjects = list(report.keys())
marks    = list(report.values())

plt.figure(figsize=(8, 4))
bars = plt.bar(subjects, marks, color="#2a9d8f")
plt.axhline(average, color="red", linestyle="--", label=f"Average = {average:.1f}")
plt.title("Student Report Card")
plt.ylabel("Marks")
plt.ylim(0, 110)

# write the value on top of each bar
for bar, mark in zip(bars, marks):
    plt.text(bar.get_x() + bar.get_width()/2, mark + 2, str(mark), ha="center")

plt.legend()
plt.show()
```
